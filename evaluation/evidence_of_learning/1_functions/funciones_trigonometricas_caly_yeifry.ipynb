## ¬øque son funciones trigonometricas?



> Es el comportamiento de la razones trigonometricas para todo los valores de $x$. Son las funciones cuyo argumento, o variable independiente, es un √°ngulo.

Definicion matematica



> Dada una circunferencia de radio $r$, al trazar dos ejes perpendiculares entre s√≠ que pasan por el origen del c√≠rculo, tendremos un sistema de ejes coordenados XY. Para calcular las diferentes funciones trigonom√©tricas, nos valdremos de estos ejes. Considerando un punto sobre la circunferencia de cordenadas (a,b) y un angulo $a$ en posicion normal asociada al punto, se define:



Sen ($a$) = $\frac{ordenada}{radio}$

* la variable $a$ debe especificarse en unidades de radianes, puesto que las funciones son con variables reales
