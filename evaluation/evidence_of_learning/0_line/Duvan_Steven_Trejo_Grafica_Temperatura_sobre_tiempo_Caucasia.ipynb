**Grafica de temperatura presente en Caucasia en base al tiempo**


```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

plt.figure(figsize = (20, 8))

Horas_del_dia_soleado_en_Caucasia = ["6:00 AM", "7:00 AM", "8:00 AM", "9:00 AM", "10:00 AM", "11:00 AM", "12:00 M", "13:00 PM", "14:00 PM", "15:00 PM", "16:00 PM", "17:00 PM", "18:00 PM", "19:00 PM", "20:00 PM", "21:00 PM", "22:00 PM", "23:00 PM"]

Temperaturas_en_grados_Celcius = [23, 25, 27, 29, 31, 33, 35, 32, 34, 35, 35, 34, 33, 32, 30, 29, 28, 27]

plt.plot(Horas_del_dia_soleado_en_Caucasia, Temperaturas_en_grados_Celcius, color = "green")

plt.grid()

```

**INFERENCIAS**

* la temperatura presente en caucasia pertenenciente en el horario de la maana (antes del meridiano) presenta un comportamiento creciente lineal con un angulo de inclinacion de 2 

* al llegar al merdiano (12 M) la temperatura llega a su maximo de 35 para acto seguido decrecer en menos de una hora 3 grados celcius (1 PM)

* tras esta reduccion notable en la temperatura, presenta una vez ms esa tendencia de subida de temperatura en base a dos grados (2) hasta llegar a la temperatura de treintaicuatro grados (34 C) para continuar con su asenso pero de una forma mucho mas "lenta" hasta llegar una vez mas a la temperatira culmen de treintaicinco grados (35 C)

* una vez llegada a la temperatura culmen este presenta un comportamiento "estable" donde no se presentan variociones en la temperatura en el lapso de dos horas (2)

* tras este momento de "estavilidad" la temperatura empieza a desplomarce de forma gradual y no cimetrica a lo largo de la horas restantes del dia hasta llegar a su punto mas bajo, que es vieticiete grados (27C) donde se mantiene estable por mas de 4  horas  (taza de decrecimiento variable)
