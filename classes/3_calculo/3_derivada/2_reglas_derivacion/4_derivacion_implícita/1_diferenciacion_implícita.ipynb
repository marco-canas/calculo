```python
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde la URL
path = '/home/marco/Documentos/docencia/groups_list/calculo_vectorial.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float) # hacer que todos los valores de la columna Parcial seen de tipo float

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.1
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Guardar los cambios en el archivo CSV
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()
    guardar_cambios()


```

# Diferenciación implícita

## Solución:  
### Metodología de solución

1. Determine $y'$ utilizando el método de diferenciación implícita
2. Halle la pendiente $m$ de la recta tangente, evaluando el punto $(2,1)$ en la función derivada $y'(x,y)$
3. Determine la recta tangente sustituyendo a $m = y'(x_{0}, y_{0}) = y'(2,1)$ y al punto de tangencia $(x_{0}, y_{0}) = (2,1)$ en la ecuación:   
   
 $$ y = m(x- x_{0}) + y_{0} $$

Este es programa en Python para graficar la curva definida por la ecuación 

$$x^2y^2 + 4xy = 12y$$

y la recta tangente a la curva en el punto $(2, 1)$.  

Utilizaremos las bibliotecas `matplotlib` y `sympy` para resolverlo.


```python


import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definimos las variables simbólicas
x, y = sp.symbols('x y')

# Definimos la ecuación de la curva
equation = x**2 * y**2 + 4 * x * y - 12 * y

# Derivadas parciales de la ecuación para obtener la pendiente de la tangente
dy_dx = sp.diff(equation, x) / sp.diff(equation, y)

# Punto de tangencia
x0, y0 = 2, 1

# Calculamos la pendiente en el punto (2, 1)
slope = dy_dx.subs({x: x0, y: y0})

# Ecuación de la recta tangente: y = m(x - x0) + y0
tangent_line = slope * (x - x0) + y0

# Convertimos la ecuación a una función de numpy para graficar
curve_func = sp.lambdify((x, y), equation, 'numpy')
tangent_func = sp.lambdify(x, tangent_line, 'numpy')

# Rango de valores para x y y
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-2, 2, 400)

# Malla para graficar la curva
X, Y = np.meshgrid(x_vals, y_vals)
Z = curve_func(X, Y)

# Graficamos la curva y la tangente
plt.figure(figsize=(8, 8))
plt.contour(X, Y, Z, levels=[0], colors='blue', linewidths=2)

# Graficamos la recta tangente
tangent_x_vals = np.linspace(-3, 3, 400)
tangent_y_vals = tangent_func(tangent_x_vals)
plt.plot(tangent_x_vals, tangent_y_vals, 'r--', \
         label=f'Recta Tangente en (2,1) y = {tangent_line}')

# Marcamos el punto de tangencia
plt.plot(x0, y0, 'go', label=f'Punto de tangencia (2, 1)')

plt.xlim(-3, 3)
plt.ylim(-2, 2)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.title('Curva $x^2y^2 + 4xy = 12y$ y su tangente en $(2,1)$')
plt.grid(True)
plt.show()

```


### Explicación del código:
1. **Definición de la ecuación**: La ecuación de la curva \(x^2y^2 + 4xy = 12y\) se define simbólicamente usando `sympy`.
2. **Derivada parcial**: Se calcula la derivada parcial con respecto a \(x\) y \(y\) para obtener la pendiente de la tangente en el punto \((2, 1)\).
3. **Ecuación de la tangente**: Usando la pendiente calculada, se define la ecuación de la recta tangente en el punto dado.
4. **Graficación**: Se grafican tanto la curva como la recta tangente utilizando `matplotlib`.

Al ejecutar el código, verás la curva junto con su recta tangente en el punto especificado.
