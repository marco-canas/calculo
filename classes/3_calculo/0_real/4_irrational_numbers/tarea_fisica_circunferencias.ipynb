a. ¿Cuántos diámetros se pueden dibujar en una circunferencia?.   

Verifique su respuesta midiendo el diámetro para todos los discos. (Tenga cuidado con las cifras significativas).

```python
import numpy as np 

import matplotlib.pyplot as plt

angulo = np.arange(0,2*np.pi+0.1, 0.1) # crear una progresión aritmética de término inicial 0 y términos final $2\pi$ y diferencia $d = 0.1$

x = np.cos(angulo) # imágens de los ángulos bajo la función coseno o grupo de abscisas. 
y = np.sin(angulo) # imágens de los ángulos bajo la función seno o grupo de ordenadas. 
plt.figure(figsize=(12, 6))
plt.plot(x, y, color = 'blue', label = 'Circunferencia unitaria')

#trazar primer diámetro
extremo1 = (-1,0)
extremo2 = (1,0)
coords_x, coords_y = zip(extremo1, extremo2)
plt.plot(coords_x, coords_y, color = 'green', label = 'primer diámetro')

#trazar segundo diámetro
extremo1 = (0,1)
extremo2 = (0,-1)
coords_x, coords_y = zip(extremo1, extremo2)
plt.plot(coords_x, coords_y, color = 'red', label = 'segundo diámetro')

#trazar tercer diámetro
extremo1 = (2**(1/2)/2, 2**(1/2)/2)
extremo2 = (-2**(1/2)/2,-2**(1/2)/2)
coords_x, coords_y = zip(extremo1, extremo2)
plt.plot(coords_x, coords_y, color = 'yellow', label = 'tercer diámetro')

plt.grid(alpha = 0.4)

plt.axis('equal')


plt.legend(loc = 'upper right') 

plt.savefig('circunferencia_con_tres_diametros.jpg')

plt.show()
```

Note que en una circunferencia se pueden trazar infinitos segmentos diametrales, uno por cada ángulo central normal (ángulo con lado inicial sobre el semieje positivo de als $x$) en la circunferencia.

¿Qué tienen de particular los resultados obtenidos?

La relación entre la circunferencia de un círculo y su diámetro es constante y se conoce como "pi" (π). El descubrimiento de esta constante se atribuye a varios matemáticos a lo largo de la historia. Uno de los primeros cálculos precisos de pi fue realizado por Arquímedes alrededor del siglo III a.C.

Arquímedes utilizó un método geométrico para aproximar el valor de pi. Posteriormente, otros matemáticos, como el matemático indio Madhava en el siglo XIV, también hicieron contribuciones significativas al cálculo de pi. Sin embargo, el símbolo π para representar esta constante fue introducido más tarde por el matemático galés William Jones en 1706 y popularizado por Leonhard Euler en el siglo XVIII. Hoy en día, pi es una constante fundamental en matemáticas y tiene un valor aproximado de 3.14159.

Esto quiere decir que la circunferencia es una variable directamente proporcional all diametro y que la constante de proporcionalidad es $\pi$

```python
import matplotlib.pyplot as plt

plt.scatter(perimetros_de_las_circunferencias, diametros, color = 'red', \
            label = 'Circunferencia_vs_diametros')

plt.plot([0, 30], [0,0], color = 'green') # trazo del eje horizontal

plt.plot([0,0], [0,9], color = 'green')

plt.grid() 

plt.legend(loc = 'upper left') 

plt.savefig('scater_circunferencia_vs_diametro.png')

plt.show()
```

* los datos están alineados
