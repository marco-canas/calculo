### Cuestionario de Prerrequisitos

1. **¿Qué es una integral definida y cómo se representa matemáticamente?**
   - **Respuesta:** Una integral definida es una operación matemática que calcula el área bajo una curva en un intervalo cerrado [a, b]. Se representa como \(\int_{a}^{b} f(x) \, dx\), donde \(f(x)\) es la función integranda, y \(a\) y \(b\) son los límites de integración.

2. **¿Qué entiendes por un sólido de revolución y cómo se genera?**
   - **Respuesta:** Un sólido de revolución es un objeto tridimensional que se obtiene al rotar una región plana alrededor de un eje. Por ejemplo, si una curva \(y = f(x)\) en el intervalo [a, b] se rota alrededor del eje x, se genera un sólido de revolución.

3. **Explica cómo se utiliza el método de discos para calcular el volumen de un sólido de revolución.**
   - **Respuesta:** El método de discos calcula el volumen de un sólido de revolución dividiendo el sólido en infinitos discos delgados perpendiculares al eje de revolución. El volumen de un disco es \(\pi \cdot [f(x)]^2 \, dx\), y sumando los volúmenes de todos los discos entre \(a\) y \(b\), obtenemos el volumen total como \(V = \int_{a}^{b} \pi [f(x)]^2 \, dx\).

# Cómo calcular volúmenes de sólidos de revolución utilizando integración definida.

[Trabajar en equipo. El trabajo de las Hormigas](https://www.youtube.com/watch?v=jW5KN4Kvpw0)

No es sorprendente que la integral definida pueda utilizarse para calcular áreas; se inventó para ese propósito. 

Pero los usos de la integral van mucho más allá de esa aplicación. 

Muchas cantidades pueden considerarse como el resultado de rebanar algo en pequeños pedazos, aproximar cada pedazo, sumarlos y tomar el límite cuando los pedazos disminuyen su tamaño. 

Este método de rebanar, aproximar e integrar puede utilizarse para encontrar los volúmenes de sólidos, siempre y cuando el volumen de cada pedazo sea fácil de aproximar.


## ¿Qué es el volumen?   

Comenzamos con sólidos sencillos denominados cilindros rectos cuatro de los cuales se muestran en la figura 1. 

En cada caso el sólido se genera moviendo una región plana (la base) a lo largo de una distancia h en dirección perpendicular a esa región. 

Y en cada caso el volumen del sólido se define como el área A de
la base por la altura $h$; esto es,

$$V = Ah$$

# El método de rebanado para determinar el volumen de un sólido al que se le conoce la función de área de la sección transversal $A(x)$. 

Ahora considere un sólido con la propiedad de que su sección transversal perpendicular a una recta dada tiene área conocida. 

En particular, supóngase que la recta es el eje $x$ y que el área de la sección transversal en $x$ es $A(x)$, $a \leq x \leq b$ (véase la figura 2).

Dividimos el intervalo $[a, b]$ insertando los puntos $a = x0 6 x1 < x2 < ··· < xn = b$.

 Después, a través de estos puntos, pasamos planos perpendiculares al eje $x$, con lo que rebanamos el sólido en capas delgadas o rebanadas (véase la figura 3). 

El volumen $\Delta V_{i}$ de una rebanada debe ser aproximadamente el volumen de un cilindro; esto es,

$$ \Delta V_{i} \approx A(\bar{x}_{i}) \Delta x_{i} $$  

(Recuérdese que $x_{i}$, denominado punto muestra, es cualquier número en el intervalo $[x_{i-1}, x_{i}]$).

El “volumen” $V$ del sólido debe estar dado, de manera aproximada, por la suma de Riemann.

$$ V \approx \sum_{i = 1}^{n} A(\bar{x}_{i}) \Delta x_{i}  $$

Cuando hacemos que la norma de la partición tienda a cero, obtenemos una integral
definida; ésta se define como el volumen del sólido.

$$ V = \int_{a}^{b} A(x)\ dx   $$

En lugar de aplicar de manera mecánica la fórmula en el recuadro para obtener
volúmenes, le sugerimos que en cada problema vaya a través del proceso que conduce
a ella. Al igual que para áreas, llamamos a este proceso rebane, aproxime, integre. Se
ilustra en los siguientes ejemplos.


# Sólidos de revolución: Método de los discos   

Cuando una región plana está por completo en un lado de una recta fija en su plano y se hace girar alrededor de esa recta, genera un sólido de revolución. 

La recta fija se denomina eje del sólido de revolución.


A manera de ilustración, si la región acotada por un semicírculo y su diámetro se
hace girar alrededor de ese diámetro, barre un sólido esférico (véase la figura 4). Si la
región dentro de un triángulo rectángulo se hace girar alrededor de uno de sus catetos,
genera un sólido cónico (véase la figura 5). Cuando una región circular se hace girar
alrededor de una recta en su plano y que no interseque al círculo (véase la figura 6), ba-
rre un toro (dona). En cada caso es posible representar el volumen como una integral
definida.

# Una región y su giro alrededor del eje $x$  

Este es un programa en Python que crea un arreglo matricial de gráfico de dimensión $1 \times 2$. En el lado izquierdo, se grafica una semicircunferencia unitaria, y en el lado derecho, se muestra la animación con `ipywidgets` de esta región girando alrededor del eje \(x\). El deslizador controla el ángulo de giro.


````python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import ipywidgets as widgets
from IPython.display import display

# Función para crear la semicircunferencia
def create_semicircle():
    theta = np.linspace(0, np.pi, 100)
    x = np.cos(theta)
    y = np.sin(theta)
    return x, y

# Función para actualizar la visualización
def update_plot(angle):
    fig = make_subplots(rows=1, cols=2, specs=[[{'type': 'scatter'}, {'type': 'surface'}]])

    # Gráfico de la semicircunferencia
    x, y = create_semicircle()
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color='blue'), name='Semicircunferencia'),
                  row=1, col=1)
    fig.add_trace(go.Scatter(x=x, y=y, fill='tozeroy', fillcolor='lightblue', mode='lines', line=dict(color='blue')),
                  row=1, col=1)
    fig.update_xaxes(title_text="x", row=1, col=1)
    fig.update_yaxes(title_text="y", row=1, col=1)

    # Crear el sólido de revolución
    theta = np.linspace(0, np.pi, 100)
    phi = np.linspace(0, 2 * np.pi, 100)
    theta, phi = np.meshgrid(theta, phi)
    x = np.cos(theta)
    y = np.sin(theta) * np.cos(phi)
    z = np.sin(theta) * np.sin(phi)
    
    fig.add_trace(go.Surface(x=x, y=y, z=z, colorscale='Blues', opacity=0.6, name='Sólido de Revolución'),
                  row=1, col=2)
    fig.update_layout(scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ), title=f"Rotación alrededor del eje x: ángulo {angle}°")

    fig.show()

# Crear el deslizador para controlar el ángulo de giro
angle_slider = widgets.IntSlider(min=0, max=360, step=1, value=0)

# Usar interact para vincular el deslizador a la función de actualización
widgets.interact(update_plot, angle=angle_slider)

# Mostrar el deslizador
display(angle_slider)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact
import ipywidgets as widgets

# Función para graficar la esfera y la semicircunferencia
def plot_sphere_and_rotate(angle):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Crear la semicircunferencia
    theta = np.linspace(0, np.pi, 100)
    x = np.cos(theta)
    y = np.sin(theta)
    
    # Generar las coordenadas 3D al girar alrededor del eje x
    phi = np.linspace(0, 2 * np.pi, 100)
    X = np.outer(x, np.ones_like(phi))
    Y = np.outer(y, np.cos(phi))
    Z = np.outer(y, np.sin(phi))

    # Graficar la superficie esférica
    ax.plot_surface(X, Y, Z, color='lightblue', alpha=0.6, rstride=5, cstride=5)

    # Graficar la curva de la semicircunferencia rotada
    angle_rad = np.radians(angle)
    X_curve = x
    Y_curve = y * np.cos(angle_rad)
    Z_curve = y * np.sin(angle_rad)
    ax.plot(X_curve, Y_curve, Z_curve, 'r', lw=2)

    # Configurar los límites de los ejes
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_zlim(-1.5, 1.5)
    ax.set_title(f"Rotación alrededor del eje x: ángulo {angle}°")
    ax.grid(True)

    plt.show()

# Utilizar interact para vincular la función de actualización con el deslizador
interact(plot_sphere_and_rotate, angle=widgets.IntSlider(min=0, max=360, step=1, value=0))
````
