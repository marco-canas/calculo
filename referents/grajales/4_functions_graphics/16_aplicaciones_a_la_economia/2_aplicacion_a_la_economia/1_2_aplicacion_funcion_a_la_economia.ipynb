## Ejemplo 16.7
Un negocio con un capital original de $20000 tiene ingresos y gastos semanales de $4000 y $3200
respectivamente. Si todas las utilidades se conservan en el negocio, exprese el valor ùëâùëâùëâùëâ del negocio al final de ùë°ùë°ùë°ùë° semanas como una funci√≥n de ùë°ùë°ùë°ùë°.  

### Soluci√≥n:  

El valor $V$ del negocio se puede expresar como la suma del valor inicial m√°s las utilidades
semanales. Como los ingresos semanales son $\$4000$ y los gastos son $\$3200$, entonces las utilidades
semanales son $4000 ‚àí 3200 = 800$. Por lo tanto, el valor del negocio despu√©s de $t$ semanas es  

$$ V = \text{Valor inicial} + \text{Utilidad semanal} \times \text{N√∫mero de semanas} $$  

Esto es,  

$$ V(t) = 20000 + 800t $$

```python
# visualizaci√≥n interactiva con plotly
import numpy as np 
import pandas as pd 
import plotly.express as px

def V(t):
    return 20_000 + 800*t

semanas = np.array([0,1,2,3,4,5,6])
valores = V(semanas)

tabla = pd.DataFrame({'t':semanas, 'V':valores})
tabla
```

```python


#trazo de diagrama de dispersi√≥n o scatter
fig = px.scatter(tabla, x = 't', y = 'V', \
                 title='Relaci√≥n entre t y V')

fig.show()

fig.write_html('relacion_entre_x_e_y.html', auto_open=True)
```
