### 1. **Función de Valor Futuro (Interés Compuesto)**
   - **Definición:**
     $$
     A(t) = P\left(1 + \frac{r}{n} \right)^{nt}
     $$
     donde:
     - $A(t)$ es el valor futuro después de $t$ años,
     - $P$ es el capital inicial,
     - $r$ es la tasa de interés anual,
     - $n$ es el número de periodos por año,
     - $t$ es el tiempo en años.
     
   - **Límite o tendencia a largo plazo:**
     Puedes analizar el límite de $A(t)$ cuando $t \to \infty$ para ver cómo el interés compuesto crece indefinidamente.
     $$
     \lim_{t \to \infty} A(t) = \infty
     $$
     Esto demuestra cómo el capital crece exponencialmente a largo plazo.

---

```python
# recurso tabular 
def elabore_tabla(f, a,b):
    import numpy as np 
    import pandas as pd 
    P,r,n = 1,0.13,12
    tiempos = np.arange(start = a, stop = b+1, step = 1)   # a_{0}, a_{n}, d
    valores_futuros = f(tiempos)
    diccionario = {'tiempo en años':tiempos, 'Valor futuro de la inversión':valores_futuros}
    tabla = pd.DataFrame(diccionario)
    tabla.to_csv('comportamiento_valor_futuro.csv', index = False)
    tabla.to_excel('comportamiento_valor_futuro.xlsx', index = False)
    return tabla
```

```python
elabore_tabla(f=lambda t:1*(1+0.1/12)**(12*t), a = 0, b = 10)
```

La lectura a la tabla, nos permite decir que a medida que el tiempo pasa, el valor del dinero se hace arbitrariamente grande sin acercarse a un valor dado. 

$$ \lim_{t \to \infty} A(t) = +\infty $$

```python
# recursoo gráfico estático
def dibujar_diagrama_dispersion(f, a, b):
    import numpy as np 
    import pandas as pd 
    import matplotlib.pyplot as plt # librería para graficos científicos 
    plt.figure(figsize = (6,4))
    plt.title('Comportamiento o tendencia del valor del dinero invertido')
    plt.xlabel('tiempo')
    plt.ylabel('Valor futuro de la inversión en millones')
    paso = 1
    dominio = np.arange(start = a, stop = b + 1, step = paso)   
    # progresiones aritméticas a_{0}, a_{n}, d
    plt.xticks(np.arange(a,b+1, step = paso)) # escalamos el eje x con una progresión aritmética
    rango = f(dominio)
    min_rango, max_rango = np.min(f(dominio)), np.max(f(dominio))
    plt.yticks(np.linspace(min_rango, max_rango, 10)) # metrizas el eje y pero con una partición regular 
    #del intervalo que va del mínimo del rango al máximo del rango
    # se mejora el escalamiento del eje y con una particion regular
    plt.axis([a-1, b+1, min_rango-1, max_rango+1]) # establecemos el rectángulo de visualización 
    plt.grid(alpha = 0.3) # dotamos al gráfico de cuadrícula para facilitar su lectura. 
    plt.scatter(dominio, rango, color = 'red', \
                label = r'Valor futuro $A(t)$', marker = '^') # 
    plt.legend() 
    # establece la convención que indica al lector qué función se esta presentando en el gráfico
    plt.savefig('tendencia_de_valor_futuro.jpg')
    plt.show()
```

```python
dibujar_diagrama_dispersion(f=lambda t:1*(1+0.1/12)**(12*t), a = 0, b = 10)
```

# Interpretaciones o descripciones del comportamiento del valor futuro a partir de la lectura del gráfico

1. El valor futuro crece a medida que el tiempo pasa
2. El valor futuro crece sin cota superior (Cuando $t \to \infty$, $A(t) \to +\infty$). Esto se nota, pues su gráfico no se está aplanando. 
3. El gráfico es concavo hacia arriba y creciente, lo que nos indica que la función tiene límite infinito en el infinito. 

```python
# recurso dinamico con ipywidgets 
def comportamiento_animado(f, a, b):
    import numpy as np
    import matplotlib.pyplot as plt
    from ipywidgets import interact 
    rastrox, rastroy = [], []
    def trazar_valor_presente(t = 0):
        plt.figure(figsize = (6,4)) 
        plt.title('Comportamiento del valor futuro a través del tiempo')
        plt.xlabel('Tiempo')
        plt.ylabel('Valor Futuro A(t)')
        min_rango = np.min(np.array([f(a), f(b)])) # la función np.min() actua con listas y array 
        max_rango = np.max([f(a), f(b)])
        plt.axis([a-1, b+1, min_rango-2, max_rango+1]) # el rectángulo de visualización debe ser definido por enteros
        plt.grid(alpha = 0.3)
        plt.xticks(np.arange(a,b))
        plt.axhline(y = 0, xmin = a, xmax = b)
        plt.axvline(x = 0, ymin = min_rango, ymax = max_rango)
        plt.scatter(t, f(t) , label = 'valor futuro', marker = '^')
        rastrox.append(t)
        rastroy.append(f(t))
        plt.scatter(rastrox, rastroy, color = 'red', marker = '^')
        plt.scatter(np.zeros_like(rastroy), rastroy, color = 'red', alpha = 0.4, \
                    marker = '^')
        plt.scatter(rastrox, np.zeros_like(rastrox), color = 'red', alpha = 0.4, \
                    marker = '>')

    interact(trazar_valor_presente, t = (a,b))    
```

```python
comportamiento_animado(f=lambda t:1*(1+0.1/12)**(12*t), a = 0, b = 20)
```

# Interpretación del gráfico dinámico



# Clase de python para constituir una clase de Python como recurso para los estudiantes para obtener representaciones tabulares y gráficas para determinar tendencias  

Aquí tienes una clase llamada `RepresentacionesTabularesGraficasTendencia` que implementa las funciones que solicitaste como métodos. La clase se puede usar para crear representaciones tabulares, gráficas estáticas y dinámicas del comportamiento de una función financiera dada, todo utilizando Python, Jupyter notebooks, y bibliotecas como `numpy`, `pandas`, `matplotlib` e `ipywidgets`.


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact

class RepresentacionesTabularesGraficasTendencia:

    # Constructor opcional para inicializar valores predeterminados si es necesario
    def __init__(self, P=1, r=0.13, n=12):
        self.P = P  # Capital inicial
        self.r = r  # Tasa de interés
        self.n = n  # Periodos de interés

    # Método para generar una tabla del comportamiento de la función
    def elabore_tabla(self, f, a, b):
        tiempos = np.arange(start=a, stop=b + 1, step=1)  # progresiones aritméticas a_0, a_n, d
        valores_futuros = f(tiempos)
        diccionario = {'tiempo en años': tiempos, 'Valor futuro de la inversión': valores_futuros}
        tabla = pd.DataFrame(diccionario)
        tabla.to_csv('comportamiento_valor_futuro.csv', index=False)
        tabla.to_excel('comportamiento_valor_futuro.xlsx', index=False)
        return tabla

    # Método para dibujar un diagrama de dispersión del comportamiento de la función
    def dibujar_diagrama_dispersion(self, f, a, b):
        paso = 1
        tiempos = np.arange(start=a, stop=b + 1, step=paso)  # progresiones aritméticas a_0, a_n, d
        valores_futuros = f(tiempos)
        plt.figure(figsize=(6, 4))
        plt.title('Comportamiento o tendencia del valor del dinero invertido')
        plt.xticks(np.arange(a, b + 1, step=paso))  # Escalar eje x con progresión aritmética
        plt.yticks(np.linspace(f(a), f(b), num=10))  # Escalar eje y con partición regular
        plt.xlabel('tiempo')
        plt.ylabel('Valor futuro de la inversión en millones')
        plt.grid(alpha=0.3)
        plt.scatter(tiempos, valores_futuros, color='red', label=r'Valor futuro $A(t)$', marker='^')
        plt.legend()
        plt.savefig('tendencia_de_valor_futuro.jpg')
        plt.show()

    # Método para generar un gráfico dinámico del comportamiento de la función
    def comportamiento_animado(self, f, a, b):
        rastrox, rastroy = [], []

        def trazar_valor_presente(t=0):
            plt.figure(figsize=(6, 4))
            plt.title('Comportamiento del valor futuro a través del tiempo')
            plt.xlabel('Tiempo')
            plt.ylabel('Valor Futuro A(t)')
            min_rango = np.min(np.array([f(a), f(b)]))
            max_rango = np.max([f(a), f(b)])
            plt.axis([a - 1, b + 1, min_rango - 2, max_rango + 1])
            plt.grid(alpha=0.3)
            plt.xticks(np.arange(a, b))
            plt.axhline(y=0, xmin=a, xmax=b)
            plt.axvline(x=0, ymin=min_rango, ymax=max_rango)
            plt.scatter(t, f(t), label='valor futuro', marker='^')
            rastrox.append(t)
            rastroy.append(f(t))
            plt.scatter(rastrox, rastroy, color='red', marker='^')
            plt.scatter(np.zeros_like(rastroy), rastroy, color='red', alpha=0.4, marker='^')
            plt.scatter(rastrox, np.zeros_like(rastrox), color='red', alpha=0.4, marker='>')

        interact(trazar_valor_presente, t=(a, b))
```

```python

# Ejemplo de uso de la clase

# Función lambda de ejemplo
f = lambda t: 1000 * np.exp(0.05 * t)  # Ejemplo de crecimiento exponencial
```

```python

# Crear una instancia de la clase
representacion = RepresentacionesTabularesGraficasTendencia()
```

```python

# Generar tabla
tabla = representacion.elabore_tabla(f, 0, 10)
print(tabla)
```

```python

# Dibujar diagrama de dispersión
representacion.dibujar_diagrama_dispersion(f, 0, 10)
```

```python

# Mostrar gráfico animado
representacion.comportamiento_animado(f, 0, 10)

```


### Descripción de los métodos:

1. **`elabore_tabla(f, a, b)`**: Este método crea una tabla con los valores de la función financiera `f` entre los años `a` y `b`. La tabla se guarda como archivo CSV y Excel.

2. **`dibujar_diagrama_dispersion(f, a, b)`**: Este método dibuja un diagrama de dispersión que muestra el comportamiento de la función `f` en el intervalo [a, b].

3. **`comportamiento_animado(f, a, b)`**: Este método genera un gráfico interactivo utilizando `ipywidgets` que permite observar cómo cambia el valor de la función `f` con el tiempo de forma dinámica.

Este código te permitirá enseñar a tus estudiantes la representación tabular y gráfica de funciones financieras y su tendencia, además de proporcionarles una visualización dinámica para reforzar su comprensión del comportamiento de las funciones en contextos económicos.

### 2. **Función de Valor Presente (Descuento)**
   - **Definición:**
     $$
     PV(t) = \frac{C}{(1 + r)^t}
     $$
     donde:
     - $PV(t)$ es el valor presente,
     - $C$ es el valor futuro o flujo de caja,
     - $r$ es la tasa de descuento,
     - $t$ es el tiempo en años.

   - **Límite o tendencia a largo plazo:**
     Explorar el límite de $PV(t)$ cuando $t \to \infty$, lo que refleja que el valor presente de un flujo de caja futuro tiende a cero a medida que el tiempo aumenta.
     $$
     \lim_{t \to \infty} PV(t) = 0
     $$
     Esto resalta la importancia del valor temporal del dinero.

---

### 3. **Función de Crecimiento Logístico en Mercados**
   - **Definición:**
     $$
     P(t) = \frac{K}{1 + \frac{K - P_0}{P_0} e^{-rt}}
     $$
     donde:
     - $P(t)$ es la cantidad en el mercado en el tiempo $t$,
     - $K$ es la capacidad de carga del mercado (máximo valor posible),
     - $P_0$ es la cantidad inicial,
     - $r$ es la tasa de crecimiento.

   - **Límite o tendencia a largo plazo:**
     El límite cuando $ t \to \infty $ muestra cómo el crecimiento de un mercado saturado se aproxima a un valor máximo $K$.

     $$
     \lim_{t \to \infty} P(t) = K
     $$
     
     Esto refleja el crecimiento limitado por la capacidad del mercado.

---



### 4. **Función de Amortización (Pago de Préstamos)**
   - **Definición:**
     \[
     A(t) = \frac{P \cdot r(1 + r)^t}{(1 + r)^t - 1}
     \]
     donde:
     - \( A(t) \) es el pago periódico,
     - \( P \) es el monto del préstamo,
     - \( r \) es la tasa de interés,
     - \( t \) es el número de pagos o periodos.

   - **Límite o tendencia:**
     Puedes analizar qué sucede con el pago periódico cuando \( t \to \infty \), es decir, qué pasa si los pagos son infinitos.
     \[
     \lim_{t \to \infty} A(t) = 0
     \]
     Esto ilustra la idea de que con el tiempo, los pagos se acercan a cero cuando la deuda está completamente pagada.

---

### 5. **Función de Costo Marginal**
   - **Definición:**
     \[
     C'(q) = \lim_{\Delta q \to 0} \frac{C(q + \Delta q) - C(q)}{\Delta q}
     \]
     donde:
     - \( C'(q) \) es el costo marginal,
     - \( C(q) \) es la función de costo total,
     - \( q \) es la cantidad de producción.

   - **Límite o tendencia:**
     Se puede analizar cómo el costo marginal se comporta a medida que \( q \to \infty \), observando cómo las economías o deseconomías de escala afectan el costo marginal.
     \[
     \lim_{q \to \infty} C'(q)
     \]
     Dependiendo de la función de costo, esto puede tender a un valor constante, indicando que el costo marginal se estabiliza con mayores niveles de producción.
